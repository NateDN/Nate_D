```python
# --------------------------------------------
# Implementação da Estrutura de FILA (FIFO)
# --------------------------------------------

TAMANHO_MAX = 25
fila = [None] * TAMANHO_MAX  # Estrutura homogênea fixa
inicio = 0
fim = 0
quantidade = 0


# --------------------------------------------
# Funções da Fila
# --------------------------------------------

def adicionar(elemento):
    global fim, quantidade
    if quantidade >= TAMANHO_MAX:
        print("Erro: Fila cheia! Não é possível adicionar mais elementos.")
    else:
        fila[fim] = elemento
        fim = (fim + 1) % TAMANHO_MAX  # Avança o índice (com volta circular)
        quantidade += 1
        print(f" '{elemento}' adicionado à fila com sucesso!")


def remover():
    global inicio, quantidade
    if quantidade == 0:
        print("Erro: Fila vazia! Nenhum elemento para remover.")
        return None
    else:
        elemento = fila[inicio]
        fila[inicio] = None  # limpa posição
        inicio = (inicio + 1) % TAMANHO_MAX
        quantidade -= 1
        print(f" '{elemento}' removido da fila com sucesso!")
        return elemento


def limpar():
    global inicio, fim, quantidade
    if quantidade == 0:
        print("A fila já está vazia.")
    else:
        for i in range(TAMANHO_MAX):
            fila[i] = None
        inicio = 0
        fim = 0
        quantidade = 0
        print("Fila completamente limpa!")


def listar():
    if quantidade == 0:
        print("A fila está vazia! Nada a listar.")
    else:
        print("\n Elementos na fila (ordem de entrada → saída):")
        idx = inicio
        for i in range(quantidade):
            print(f"[{i+1}] {fila[idx]}")
            idx = (idx + 1) % TAMANHO_MAX


def vazia():
    return quantidade == 0


# --------------------------------------------
# Menu principal da Fila
# --------------------------------------------

def menu():
    while True:
        print("\n===== MENU DA FILA =====")
        print("1 - Adicionar elemento à fila")
        print("2 - Remover elemento da fila")
        print("3 - Listar elementos")
        print("4 - Limpar fila")
        print("5 - Verificar se a fila está vazia")
        print("0 - Sair")
        opcao = input("Escolha uma opção: ").strip()

        if opcao == "1":
            elemento = input("Digite o elemento para adicionar: ").strip()
            if elemento:
                adicionar(elemento)
            else:
                print("Valor vazio não pode ser adicionado.")
        elif opcao == "2":
            remover()
        elif opcao == "3":
            listar()
        elif opcao == "4":
            limpar()
        elif opcao == "5":
            if vazia():
                print("A fila está vazia.")
            else:
                print(f"A fila contém {quantidade} elemento(s).")
        elif opcao == "0":
            print("Saindo do sistema...")
            break
        else:
            print("Opção inválida. Tente novamente.")


# --------------------------------------------
# Execução
# --------------------------------------------

if __name__ == "__main__":
    try:
        menu()
    except KeyboardInterrupt:
        print("\n Programa interrompido pelo usuário.")
```
