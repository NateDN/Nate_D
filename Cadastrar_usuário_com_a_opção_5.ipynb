```python
# Estruturas homog√™neas
nomes = ["Ana", "Carlos", "Beatriz", "Daniel", "Fernanda"]
idades = [25, 17, 30, 19, 40]
habilitacoes = [True, False, True, False, True]

limite = 50

def listar_usuarios():
    print("\n----- Usu√°rios Cadastrados -----")
    for i in range(len(nomes)):
        hab = "Sim" if habilitacoes[i] else "N√£o"
        print(f"{i+1} - {nomes[i]} | Idade: {idades[i]} | Habilita√ß√£o: {hab}")

def validar_dirigir():
    nome_busca = input("Digite o nome para validar permiss√£o: ").strip()
    if nome_busca in nomes:
        i = nomes.index(nome_busca)
        if idades[i] >= 18 and habilitacoes[i]:
            print(f"‚úÖ {nomes[i]} PODE dirigir.")
        else:
            print(f"‚ùå {nomes[i]} N√ÉO pode dirigir.")
    else:
        print("‚ö†Ô∏è Usu√°rio n√£o encontrado.")

def buscar_usuario():
    nome_busca = input("Digite qualquer parte do nome que deseja buscar: ").strip().lower()
    encontrado = False
    for i in range(len(nomes)):
        if nome_busca in nomes[i].lower():
            print("\nüéâ Usu√°rio encontrado!")
            print(f"Posi√ß√£o: {i+1}")
            print(f"Nome: {nomes[i]}")
            print(f"Idade: {idades[i]}")
            print(f"Habilita√ß√£o: {'Sim' if habilitacoes[i] else 'N√£o'}")
            print("-------------------------")
            encontrado = True
    if not encontrado:
        print("‚ùå Nenhum usu√°rio encontrado com esse crit√©rio.")

def remover_usuario():
    nome_busca = input("Digite o nome exato do usu√°rio a ser removido: ").strip()
    if nome_busca in nomes:
        i = nomes.index(nome_busca)
        # Remove o usu√°rio das 3 listas
        nomes.pop(i)
        idades.pop(i)
        habilitacoes.pop(i)
        print(f"‚úÖ Usu√°rio '{nome_busca}' removido com sucesso!")
    else:
        print("‚ùå Usu√°rio n√£o encontrado para remo√ß√£o.")

def cadastrar_usuario():
    if len(nomes) < limite:
        nome = input("Digite o nome: ").strip()
        idade = int(input("Digite a idade: "))
        resp = input("Possui habilita√ß√£o? (s/n): ").strip().upper()
        habilitado = True if resp == "S" else False

        nomes.append(nome)
        idades.append(idade)
        habilitacoes.append(habilitado)
        print("‚úÖ Usu√°rio cadastrado com sucesso!")
    else:
        print("‚ö†Ô∏è Limite de cadastros atingido.")

# --------- MENU PRINCIPAL ---------
while True:
    print("\n===== MENU =====")
    print("1 - Cadastrar novo usu√°rio")
    print("2 - Listar todos os usu√°rios")
    print("3 - Validar permiss√£o para dirigir")
    print("4 - Buscar usu√°rio pelo nome")
    print("5 - Remover usu√°rio")
    print("0 - Sair")
    opcao = input("Escolha uma op√ß√£o: ")

    if opcao == "1":
        cadastrar_usuario()
    elif opcao == "2":
        listar_usuarios()
    elif opcao == "3":
        validar_dirigir()
    elif opcao == "4":
        buscar_usuario()
    elif opcao == "5":
        remover_usuario()
    elif opcao == "0":
        print("Saindo do sistema... At√© logo!")
        break
    else:
        print("‚ö†Ô∏è Op√ß√£o inv√°lida. Tente novamente.")
```
